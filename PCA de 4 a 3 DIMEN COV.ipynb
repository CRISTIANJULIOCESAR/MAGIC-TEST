```python
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np


#data = datasets.load_digits()
data = datasets.load_iris()
X = data.data
y = data.target

# Mean centering
mean = np.mean(X)
X = X - mean
# covariance, function needs samples as columns
cov = np.cov(X.T)

# eigenvalues, eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(cov)
# -> eigenvector v = [:,i] column vector, transpose for easier calculations


# sort eigenvectors
eigenvectors = eigenvectors.T
eigenvectors

idxs = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[idxs]
eigenvectors = eigenvectors[idxs]


n_components =3

# store first n eigenvectors
components = eigenvectors[0:n_components]
#reducí la dimencionalidad a 2 componentes
components 
X_projected =np.dot(X,components.T)
```

```python
print('Shape of X:', X.shape)
print('Shape of transformed X:', X_projected.shape)

```

```python
import pandas as pd

scores_df = pd.DataFrame(X_projected, columns=['PC1', 'PC2','PC3'], index = None)
scores_df
```

```python
#Método para retomar mis etiquetas  de especies 
Y_label = []

for i in y:
    if i == 0:
        Y_label.append('Setosa')
    elif i == 1:
        Y_label.append('Versicolor')
    else:
        Y_label.append('Virginica')

Species = pd.DataFrame(Y_label, columns=['Species'])
```

```python
df_scores = pd.concat([scores_df, Species], axis=1)
```

```python
import plotly.express as px

fig = px.scatter_3d(df_scores, x='PC1', y='PC2', z='PC3',
              color='Species')

fig.show()
```
