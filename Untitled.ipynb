```python
import magic
import scprep

import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

# Matplotlib command for Jupyter notebooks only
%matplotlib inline


```

```python


emt_data = scprep.io.load_csv('HMLE_TGFb_day_8_10.csv.gz', cell_names=False)

# Remove empty columns and rows
#emt_data = scprep.filter.remove_empty_cells(emt_data)

#emt_data = scprep.filter.remove_empty_genes(emt_data)

# Filter by library size to remove background
scprep.plot.plot_library_size(emt_data, cutoff=1500)
emt_data.head()
```

```python
emt_data = scprep.normalize.library_size_normalize(emt_data)
# be sure to uncomment this, unless your data is pre-transformed
emt_data = scprep.transform.sqrt(emt_data)
emt_data.head()
```

Normalizando sus datos

Después de filtrar, los siguientes pasos son realizar la normalización y transformación del tamaño de la biblioteca. La transformación logarítmica se usa con frecuencia para la secuencia de ARN de una sola célula, sin embargo, esto requiere la adición de un pseudocuenta para evitar valores infinitos en cero. En su lugar, usamos una transformación de raíz cuadrada, que tiene propiedades similares a la transformación de registro pero no tiene ningún problema con los ceros.

# Ejecutando MAGIC

Ahora que sus datos han sido preprocesados, está listo para ejecutar MAGIC.

# Creando el operador MAGIC

Si no especifica parámetros, MAGIC crea un operador con los siguientes valores predeterminados: knn = 5, knn_max = 3 * knn, decay = 1, t = 3.

```python
magic_op = magic.MAGIC( t=3, knn = 10, n_pca=20)
```

Ejecutando MAGIC con selección de genes

La función magic_op.fit_transform toma los datos normalizados y una matriz de genes seleccionados como argumentos. Si no se proporcionan genes, MAGIC devolverá una matriz de todos los genes. Lo mismo puede lograrse sustituyendo la matriz de nombres de genes con genes = 'all_genes'.

```python
emt_magic = magic_op.fit_transform(emt_data, genes=['VIM', 'ZNF217', 'ZEB1'])
```

Visualizando relaciones gen-gen

Podemos ver las relaciones gen-gen mucho más claramente después de aplicar MAGIC. Tenga en cuenta que el cambio en los valores absolutos de la expresión génica no es significativo: la diferencia relativa es lo único que importa.

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(16, 6))


scprep.plot.scatter(x=emt_magic['VIM'], y=emt_magic['ZEB1'], c=emt_magic['ZNF217'], ax=ax2,
                    xlabel='VIM', ylabel='ZEB1', legend_title="ZNF217", title='After MAGIC')

scprep.plot.scatter(x=emt_data['VIM'], y=emt_data['ZEB1'], c=emt_data['ZNF217'],  ax=ax1,
                    xlabel='VIM', ylabel='ZEB1', legend_title="ZNF217", title='Before MAGIC')

plt.tight_layout()
plt.show()
```
